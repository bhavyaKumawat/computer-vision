Import necessary Libraries

```python
from tensorflow.keras import datasets
import matplotlib.pyplot as plt
from keras.utils import load_img, img_to_array
```

Download Dataset

```python
help(datasets.fashion_mnist.load_data)
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
```



---




### Visualize_Numpy_Array

```python
img = train_images[0]
type(img), img.shape
```

```python
plt.imshow(img) # Display (NumPy array/PIL image) as an image
plt.axis('off') #remove axis lines, tick marks, and labels
plt.show() #Display all open figures
```

Adjust the figure size

```python
plt.figure(figsize=(1, 1))  # Set width=8 inches and height=6 inches
plt.imshow(img)
plt.axis('off')
plt.show()
```



---




### Visualize_multiple_Numpy_Arrays

```python
plt.figure(figsize=(5,5)) #plt.figure(): Creates a new figure
for i in range(25):
    plt.subplot(5, 5, i+1) #Add an Axes to the current figure (nrows, ncols, index). index starts at 1
    plt.axis('off')
    plt.imshow(train_images[i])
plt.show()
```

Provide a description

**Approach 1 :**

```python
plt.figure(figsize=(5,6))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([]) #removes all xticks
    plt.yticks([]) #removes all yticks
    plt.imshow(train_images[i])
    plt.xlabel(train_labels[i]) # set x-axis label
plt.show()
```

**Approach 2 :**

```python
plt.figure(figsize=(5,6))
plt.suptitle("MNIST") #set the title for the entire figure
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i])
    plt.title(train_labels[i]) #set the title of a plot
plt.show()
```

### Visualize_multiple_Images_in_a_Folder

```python
import glob
from matplotlib.image import imread
dir = "birds"
files = glob.glob(f"{dir}/*") #returns paths matching a pattern
```

```python
plt.figure(figsize=(8,2)) 
for i in range(3):
  img = imread(files[i])
  plt.subplot(1, 3, i+1)
  plt.axis('off')
  plt.imshow(img)
plt.show()
```
